# Maschinelles Lernen

> Maschinen lernen nicht. 

Ok. Was ist dann Maschinelles Lernen?

## Mathematische Problembeschreibung

Wir werden uns mit Daten beschäftigen, die als *Matrizen* bzw. *Tabellen* gespeichert sind.

Zum Beispiel:

* Erfasse von $N=100$ Personen $p=2$ Eigenschaften: Größe und Gewicht.
* Daraus ergibt sich eine $N \times p$ Matrix $\m{X}$
* $\v{X}_0$ ist die erste Zeile der Tabelle. Größe und Gewicht der ersten Person
* $\v{X}_{\bullet 1}$ ist die zweite Spalte der Tabelle. Das Gewicht aller Personen.


Eine Aufgabe könnte nun sein, einen Algorithmus zu finden, der aus diesen Daten das Alter der Personen schätzt.

* Bezeichnen wir diese Zielgröße mit $\m{Y}$, das Alter aller Personen.


Eine mögliche Definition von Maschinellem Lernen ist dann

> Gegeben die $N \times p$ Matrix $\m{X}$ und einen dazugehörigen Ziel-Vektor $\m{Y}$:  
> Finde eine Funktionsvorschrift $f(X) = \hat{Y}$, die eine Beobachtung $X$ als Eingabe bekommt und
eine Schätzung $\hat{Y}$ zurückgibt,  
> wobei eine sinnvoll gewählte „Verlustfunktion“ $L(Y, f(X))$ minimiert wird für $\m{X}$.

__Unsere Aufgabe__

Gegeben $\boldsymbol{X} = (\p{Ticketgebühr}, \p{Klasse}, \p{Alter}, \p{Geschlecht})$,  
können wir eine Funktion finden, die vorhersagt, ob ein bestimmter Passagier das Unglück überlebt hat?


Dazu müssen wir uns erst einmal überlegen, wann eine solche Funktion "gut" ist.

Wir berechnen ein paar Kennzahlen für jede Funktion:

| | | |
|---|---|---|
|__True Positives__  | $\TP$ | Anzahl der in Wahrheit Überlebenden, die die Funktion korrekt vorhergesagt hat |
|__False Positives__ | $\FP$ | Anzahl der in Wahrheit Gestorbenen, die die Funktion falsch vorhergesagt hat |
| __True Negatives__  | $\TN$ | Anzahl der in Wahrheit Gestorbenen, die die Funktion korrekt vorhergesagt hat |
| __False Negatives__ | $\FN$ | Anzahl der in Wahrheit Überlebenden, die die Funktion falsch vorhergesagt hat |


Zusammen kann man den Anteil der korrekt vorgesagten Passagiere berechnen:


$$
    \operatorname{Genauigkeit}(\v{y}, \v{\hat{y}}) =  \frac{\TP + \TN}{ \TP + \FP + \FN + \TN} = \frac{\text{Anzahl korrekt vorgesagter Passagiere}}{\text{Gesamtanzahl Passagiere}}.
$$


Unsere Basis wird sein, immer den häufigsten Wert, also "Gestorben" zu raten.

Eine Funktion muss besser als diese Basis sein, um einen Mehrwert zu haben

```python
truth = passengers['survived']
```

__Die Basis__:

Wir raten für jeden Passagier den häufigsten Wert (`False` = Gestorben)

```python
def predict(passengers):
    '''Everyone dies'''
    return pd.Series(False, index=passengers.index)


prediction = predict(passengers)

plotting.plot_bars_and_confusion(truth=truth, prediction=prediction)
```

__Eine etwas clevere Lösung__:

Auf der Titanic galt vielleicht "Frauen und Kinder zuerst"?

```python
def predict(passengers):
    '''Women and children first'''
    # | means "or"
    return (passengers['sex'] == 'female') | (passengers['age'] <= 12)

prediction = predict(passengers)

plotting.plot_bars_and_confusion(truth=truth, prediction=prediction)
```

### Alternative Hypothese: Reiche zuerst



```python
def predict(passengers):
    '''People in first class survived'''
    return passengers['pclass'] == 1

prediction = predict(passengers)

plotting.plot_bars_and_confusion(truth=truth, prediction=prediction)
```

# Anwendung von Maschinellen Lernen


## Lineare Modelle


Lineare Modelle kombinieren alle Attribute unseres Datensatzes in der Form:


$$
f(x) = a + b \cdot X_0 + c \cdot X_1 + d \cdot X_2 + ....
$$

Also zum Beispiel für unseren Fall hier:

$$
f(x) = a + b \cdot \p{Ticketgebühr} + c \cdot \p{Klasse} + d \cdot \p{Alter} + e \cdot \p{Geschlecht})
$$


Die Aufgabe des Maschinellen Lernens ist es nun die Parameter $a, b, c, d, e$ zu finden, die am besten unser Problem lösen.

Hierfür gibt es tatsächlich eine analytische Lösung, die aus ein paar Matrizenprodukten und Invertierungen der Eingangsdaten steht.

Die Phase, in der wir Parameter unseres Modells bestimmen, nennen wir "Training". Hier findet das eigentliche Lernen statt.

Später können wir das gelernte Modell auf neue, unbekannte Daten anwenden und Vorhersagen machen.

Für das Training braucht man einen Datensatz, auf dem sowohl die Attribute ($\m{X}$) als auch die Zielgröße ($\v{Y}$) bekannt sind.

Deshalb wird dieses Vorgehen auch "überwachtes Lernen" genannt.

Oft ist es schwierig und aufwendig, einen solchen Trainingsdatensatz für das aktuelle Problem zu bekommen:

<img src="https://www.keferstein-ev.de/mitgliederbereich/blog/wp-content/uploads/2019/11/google_recaptcha_v3_ampel.png">

## Lineare Regression für den Titanic Datensatz



```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split

features = [
    'sex_code',
    'pclass',
    'fare',
    'age',
]

target = 'survived'

passengers[features + [target]]
```

Um die Güte unseres Modell zu bestimmen, ist es wichtig, die Genauigkeit auf Daten
zu bestimmen, die unser Model noch nicht kennt.

Alles andere wäre geschummelt ;-)

```python
# split data in indepent sets for training / validation
passengers_train, passengers_test = train_test_split(
    passengers,
    test_size=0.5, # 50/50 split train / test
    # keep proportions of survived / not-survived the same in the split datasets
    stratify=passengers[target],
    random_state=0,
)

len(passengers_train), len(passengers_test)
```

Wir trainieren das Modell auf dem Trainings-Datensatz:

```python
# fitting the model means calculating the parameters a, b, c, ...
model = linear_model.LinearRegression()
model.fit(passengers_train[features], passengers_train[target])

print('a={:.2f}, b={:.2f}, c={:.2f}, d={:.4f}, e={:.4f}'.format(model.intercept_, *model.coef_))
```

Und wenden es auf den Test-Datensatz and und berechnen die Güte:

```python
# for a linear model, don't get 0 or 1 but a real value
prediction = model.predict(passengers_test[features]) > 0.5

plotting.plot_bars_and_confusion(truth=passengers_test[target], prediction=prediction)
```

Ein kleines bisschen besser.

(vielleicht, wir müssten auch Unsicherheiten der Güte bestimmen um das wirklich mit Sicherheit sagen zu können)

## Entscheidungsbäume


Entscheidungsbäume sind ein weiterer Algorithmus zur Klassifizierung und lassen sich intuitiv nachvollziehen.

Der Trainingsprozess ist allerdings mathematisch aufwendinger, darum heute außen vor.

Die Idee ist immer wieder einfache Entscheidungen zu treffen und so den Datensatz anhand der Attribute immer weiter aufzuteilen.

Bis der Datensatz irgendwann so unterteilt ist, das die einzelnen Untergruppen nur noch eine der Zielklassen enthalten.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from IPython.display import Image
```

```python
model = DecisionTreeClassifier(max_depth=3, criterion='entropy')

model.fit(passengers_train[features], passengers_train[target])


Image(plotting.draw_tree(
    model,
    feature_names=features,
    class_names=['Überlebt', 'Gestorben'],
    fontname='Fira Sans',
    colors=['xkcd:light blue', 'xkcd:light red']
))
```

```python
prediction = model.predict(passengers_test[features])

plotting.plot_bars_and_confusion(truth=passengers_test[target], prediction=prediction)
```

## Random Forests

Zum Abschluss, noch ein in der Physik sehr beliebter Algorithmus: der Random Forest.

Ein Random Forest ist ein Ensemble aus vielen Entscheidungsbäumen.

Damit nicht alle Bäume gleich sind, wird
* Jeder Baum auf einem Datensatz trainiert, der zufällig aus dem eigentlichen Datensatz generiert wird
* Darf jeder Baum bei jeder "Abfrage" nicht alle Attribute angucken, sondern nur eine zufällige Untermenge

Wir bilden den Mittelwert der Vorhersagen aller Bäume

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, criterion="entropy", min_samples_leaf=10, random_state=0)
model.fit(passengers_train[features], passengers_train[target])

Image(plotting.draw_tree(
    model.estimators_[50],
    feature_names=features,
    class_names=['Überlebt', 'Gestorben'],
    fontname='Fira Sans',
))
```

```python
prediction = model.predict(passengers_test[features])

plotting.plot_bars_and_confusion(truth=passengers_test[target], prediction=prediction)
```
